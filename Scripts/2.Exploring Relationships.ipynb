## 6. Creating categorical plot

```python
sns.histplot(df['PM2.5 AQI Value'], bins = 20, kde = True)
```

```python
sns.lmplot(data=df,y='PM2.5 AQI Value',x='AQI Value',hue="AQI Category")
```

```python
sns.lmplot(data=df,x='CO AQI Value',y='NO2 AQI Value',hue="AQI Category")
```

```python
aql_overall = df[['Country','AQI Value','AQI Category']]
aql_overall_haz = aql_overall[aql_overall['AQI Category'] == 'Hazardous']
hazardous = aql_overall_haz.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value',ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=hazardous)
plt.title('Worst Air Quality Countries Based on Overall AQI')
```

```python
aql_overall_good = aql_overall[aql_overall['AQI Category'] == 'Good']
good = aql_overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value').head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=good)
plt.title('High Air Quality Countries Based on Overall AQI')
```

```python
aql_overall_good = aql_overall[aql_overall['AQI Category'] == 'Unhealthy for Sensitive Groups']
good = aql_overall_good.groupby('Country',as_index=False)['AQI Value'].mean().sort_values(by='AQI Value',ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x='AQI Value', y='Country',data=good)
plt.title('High Air Quality Countries Based on Overall AQI')
```

#### Rusia, USA, and India are top 3 countries that have bad air qualiy.
