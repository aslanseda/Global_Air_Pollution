### 3.4. Outliers

```python
# Calculating quartiles
q1 = df['AQI Value'].quantile(0.25)
q2 = df['AQI Value'].quantile(0.50)
q3 = df['AQI Value'].quantile(0.75)
```

```python
std_dev = df['AQI Value'].std()
mean = df['AQI Value'].mean()
lower_bound = mean - (2 * std_dev)
upper_bound = mean + (2 * std_dev)
```

```python
#Defining outliers 
outliers_1 = df[(df['AQI Value'] < lower_bound) | (df['AQI Value'] > upper_bound)]
```

## 4. Consistency Checks

### 4.1 Checking for mixed-type data columns

```python
for col in df.columns.tolist():
  weird = (df[[col]].applymap(type) != df[[col]].iloc[0].apply(type)).any(axis = 1)
  if len (df[weird]) > 0:
    print (col, ' mixed')
  else: 
    print(col, ' consistent')
```
